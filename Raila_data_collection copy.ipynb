```python
from googleapiclient.discovery import build
import pandas as pd

api_key = "AIzaSyCFZZR6iyblZdVIbl85BwUec5wR8v6KBKU"

youtube = build("youtube", "v3", developerKey=api_key)

query = "Raila Odinga death"

# Step 1: Search top 50 videos
search_response = youtube.search().list(
    q=query,
    part="id,snippet",
    maxResults=50,
    type="video",
    order="viewCount"
).execute()

videos = []

# Step 2: Collect statistics for each video
for item in search_response['items']:
    video_id = item['id']['videoId']
    title = item['snippet']['title']

    stats_response = youtube.videos().list(
        part="statistics",
        id=video_id
    ).execute()

    # Some videos may not have all stats  handle safely
    if stats_response["items"]:
        stats = stats_response["items"][0]["statistics"]
        videos.append({
            "video_id": video_id,
            "title": title,
            "viewCount": int(stats.get("viewCount", 0)),
            "likeCount": int(stats.get("likeCount", 0)),
            "commentCount": int(stats.get("commentCount", 0))
        })

# Step 3: Convert to DataFrame
df_videos = pd.DataFrame(videos)

print(" Successfully created df_videos with these columns:")
print(df_videos.columns)
df_videos.head()
```

```python
from googleapiclient.errors import HttpError

all_comments = []

for video_id in df_videos['video_id'].head(20):  # Top 20 videos
    print(f" Fetching comments for video: {video_id}")
    comments = []
    next_page_token = None
    total_fetched = 0

    try:
        while total_fetched < 20:  # limit to 20 comments per video
            response = youtube.commentThreads().list(
                part="snippet",
                videoId=video_id,
                maxResults=50,
                order="relevance",  # top comments first
                textFormat="plainText",
                pageToken=next_page_token
            ).execute()

            for item in response["items"]:
                snippet = item["snippet"]["topLevelComment"]["snippet"]
                comments.append({
                    "video_id": video_id,
                    "author": snippet.get("authorDisplayName"),
                    "comment": snippet.get("textDisplay"),
                    "likeCount": snippet.get("likeCount", 0)
                })
                total_fetched += 1
                if total_fetched >= 20:
                    break

            next_page_token = response.get("nextPageToken")
            if not next_page_token:
                break

    except HttpError as e:
        # Handle disabled comments or restricted access
        if e.resp.status == 403 and "commentsDisabled" in str(e):
            print(f" Comments disabled for video: {video_id}. Skipping...")
        else:
            print(f" Error fetching comments for video {video_id}: {e}")
        continue

    all_comments.extend(comments)

# Convert to DataFrame
df_comments = pd.DataFrame(all_comments)
print(" Successfully fetched comments!")
print(df_comments.head())
```

```python
from textblob import TextBlob

# Drop missing comments
df_comments = df_comments.dropna(subset=["comment"])

# Apply sentiment analysis
def get_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity  # -1 (negative)  +1 (positive)
    if polarity > 0.1:
        return "Positive"
    elif polarity < -0.1:
        return "Negative"
    else:
        return "Neutral"

df_comments["sentiment"] = df_comments["comment"].apply(get_sentiment)

# Count results
sentiment_summary = df_comments["sentiment"].value_counts(normalize=True) * 100
print(" Sentiment Breakdown (%):")
print(sentiment_summary.round(2))
```

```python
import matplotlib.pyplot as plt

sentiment_summary.plot(kind='bar', figsize=(6,4))
plt.title("Public Sentiment on Raila's Death (YouTube Comments)")
plt.xlabel("Sentiment")
plt.ylabel("Percentage")
plt.xticks(rotation=0)
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
import pandas as pd

#  Clean your comments
df_comments["clean_comment"] = df_comments["comment"].str.lower().str.replace(r"[^a-zA-Z\s]", "", regex=True)

#  TF-IDF Vectorizer (you can tweak max_features for larger datasets)
vectorizer = TfidfVectorizer(stop_words='english', max_features=500)
X = vectorizer.fit_transform(df_comments["clean_comment"])

#  Use NMF to identify topics
nmf = NMF(n_components=5, random_state=42)
nmf.fit(X)

#  Extract top words per topic
feature_names = vectorizer.get_feature_names_out()
num_top_words = 10

for i, topic in enumerate(nmf.components_):
    print(f"\n Topic {i+1}:")
    top_words = [feature_names[j] for j in topic.argsort()[:-num_top_words - 1:-1]]
    print(", ".join(top_words))
```
